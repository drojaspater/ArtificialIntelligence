## <center> ¿Cómo lo hace? </center>

<center> La pagina es un modelo de una red neuronal realizado con la libreria  TensorFlow, el repositorio de github con el desarrollo del modelo es https://github.com/tensorflow/playground  </center>

## <center> Opciones dentro de Tenserflow - Playground </center>
Al ser una red interactiva interactiva, tendremos acceso a todos los parametros de entrenamiento de la red, dentro de los cuales se encuentran:

1. **Capa de entrada:** Esto nos permitira tender control sobre las neuronas de entrada y el tipo de neurona que es, de modo que podremos escoger entre neuronas lineales, cuadraticas y sinusoidales.
2. **Capas ocultas:** Podremos escoger el número de capas ocultas (Máximo 6) y la cantidad de neuronas en cada capa.
3. **Datos de entrada:** Los cuales son los datos con los cuales haremos el entrenamiento y la prueba del modelo.
4. **Tasa de aprendizaje:** De modo que tendremos control sobre las variaciones a los parametros en cada epoca.
![Tasa de aprendizaje](https://miro.medium.com/v2/resize:fit:1400/0*J8Gsz4498OOwD1Xd.png)
5. **Función de activación:** Permitiendos escoger entre 4 funciones de activación diferentes las cuales tienen la tarea de cambiar la información de salida de cada neurona.
![fución de activación](https://miro.medium.com/v2/resize:fit:640/format:webp/1*ifhhhlve3jETpsfX6TOQmw.png)
6. **Función de Regularización y tasa de regularización** La cual es una función que nos permite adicionar el peso de los parametros a la función de activación con el objetivo de evitar un sobre ajuste y se controlan dichos aportes con la tasa de regularización.

7. **Proporción de entrenamiento a datos de prueba:** Permitiendonos escoger la proporción de los datos de entrada que seran utilizados como datos de entrenamiento y la proporción que seran utilizados como datos de prueba.
![](https://h8v7k6i3.rocketcdn.me/wp-content/uploads/2022/02/tvt.jpg)
8. **Noise:** Corresponde al ruido sobre la distrubución de datos iniciales.
9. **Mini Batch Size:** El cual nos permite escoger el tamaño del lote para cada epoca. 

con esto, la pagina nos retornara tres cosas esenciales para comprender el funcionamiento de la red neuronal simulada:
1. **Datos de salida:** Mostrandonos los datos de entrada en un diagrama de color de dos dimensiones y a la par las predicciones realizadas por el modelo.
2. **Epocas:** De modo que sabemos que tan rapido se entrena el modelo y tenemos control sobre ellas como una variable temporal.
3. **Función de perdida:** La cual nos permite ver si el modelo ya realizo el entrenamiento y el error con respecto a los datos de prueba.
